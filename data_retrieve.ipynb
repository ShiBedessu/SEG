```python
from dotenv import load_dotenv
import os
import mysql.connector
import pandas as pd
```

```python


# Nạp biến môi trường từ file .env
load_dotenv(dotenv_path="C:\Code\Ky5\SEG\Crawl\database\.env")

host = os.getenv("MYSQL_HOST")
user = os.getenv("MYSQL_USER")
password = os.getenv("MYSQL_PASSWORD")
database = os.getenv("MYSQL_DB")

# Kết nối tới MySQL
conn = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database
)

# Viết truy vấn SQL
query = "SELECT * FROM law_chunks Order By auto_id ASC LIMIT 10"

# Đọc dữ liệu vào DataFrame
df = pd.read_sql(query, conn)

# Hiển thị dữ liệu
print(df.head())

# Đóng kết nối
conn.close()
```
